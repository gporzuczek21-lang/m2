```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt
cena_teraz = 120000
inflacja = 0.05
ilosc_miesiecy_w_roku = 12
wspolczninik_wzrostu = inflacja / ilosc_miesiecy_w_roku
okres_lat = 5
ilosc_miesiecy = okres_lat * ilosc_miesiecy_w_roku
cena_w_czasie = []
oprocenotowanie_na_lokacie = 0.12
oprocenotowanie = oprocenotowanie_na_lokacie / ilosc_miesiecy_w_roku 
pv=  0
if __name__=="__main__":
    peridos = np.arange(1,ilosc_miesiecy, dtype=int)
    for i in (peridos):
        a = cena_teraz*((1+wspolczninik_wzrostu)**i)
        a = round(a,2)
        cena_w_czasie.append(a)

    cena = np.array(cena_w_czasie)
    fv = cena[-1]
    

    wplata = (-npf.pmt(oprocenotowanie,ilosc_miesiecy,pv,fv))
    wplata = round(wplata,2) 

    suma_kapitalu =[wplata]
    pierwsza = wplata*(1+oprocenotowanie)
    
    for p in (peridos):
        p =(pierwsza+wplata)*(1+oprocenotowanie)
        pierwsza = round(p,2)
        suma_kapitalu.append(pierwsza)

    lokata = np.array(suma_kapitalu) 
  


    print(f"za {okres_lat} cena mieszkania bedzie wynosiła {fv}zł")
    print(f"Na lokate w banku nalezy co miesią wpłacać {wplata}zł")
    plt.plot(lokata,label='wartosc kapiatłu')   
    plt.plot(cena,label='cena mieszkania')
    plt.legend()
    plt.xlabel('Liczba okresów')
    plt.ylabel('waluta "zł"')
    plt.show
```
